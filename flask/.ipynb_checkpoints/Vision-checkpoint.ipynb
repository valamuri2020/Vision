```python
count = df_mock['UNITID'].count()
# KMC on mock
# find centers
# corrwith those centers with df_further_reduced
```

```python
cols = "ACTCMMID ACTENMID ACTMTMID ACT_AVG ADM_RATE_ALL SAT_AVG_ALL".split()
df_mock_reduced = df_mock[cols]
df_mock_reduced.head()
```

```python
def k_value(row_count):
    if row_count > 0 and row_count < 5:
        return 1
    else:
        return 5
```

```python
from sklearn.cluster import KMeans
```

```python
kmeans = KMeans(n_clusters = k_value(count))
```

```python
kmeans.fit(df_mock_reduced)
```

```python
kmeans.cluster_centers_
```

```python
cluster_center = pd.DataFrame(kmeans.cluster_centers_, columns=cols)
```
